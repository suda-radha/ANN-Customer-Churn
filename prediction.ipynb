```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## Load teh trained model, scalar pickle file, onehotendoded
#model=load_model('model.h5', compile=False)
model=load_model('model_saved.keras', compile=False)

## load the encoder and scalar
with open('onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender=pickle.load(file)

with open('scalar.pkl', 'rb') as file:
    scalar = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore' : 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard':1,
    'IsActiveMember': 1,
    'EstimatedSalary':50000
}
```

```python
# One-hot encode 'Geography'
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]])
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
3# Encode categorical variable - Gender
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## concatenation one hot encoded - geography to input_df
input_df = pd.concat([input_df.drop("Geography", axis=1),geo_encoded_df], axis=1)
input_df
```

```python
# scaling the input data - now u scaled teh input data (all values scaled between-1 to 1)
input_scaled=scalar.transform(input_df)
input_scaled
```

```python
## Predict churn
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0] 
prediction_proba
```

```python
if prediction_proba>0.5:
    print('The customer is likely to Churn.')
else:
    print('The customer is Not likely to churn.')
```
